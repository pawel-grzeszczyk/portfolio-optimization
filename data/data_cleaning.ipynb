```python
import pandas as pd
```

```python
# Reading data, renaming columns and reordering
w20_stock_close = pd.read_csv('./data_files/w20_stock.csv').filter(regex='^Close_')
w20_stock_close.columns = [x.split('_')[1].split('.')[0] for x in w20_stock_close.columns]
w20_stock_close['Date'] = pd.read_csv('./data_files/w20_stock.csv')['Date']
w20_stock_close.insert(0, 'Date', w20_stock_close.pop('Date'))

# Close_PCO.WA limits data range
w20_stock_close = w20_stock_close.loc[w20_stock_close['Date'] > '2021-05-26'].reset_index(drop=True)
w20_stock_close.head()
```

```python
for stock in w20_stock_close.columns.drop('Date'):
    w20_stock_close[stock] = (w20_stock_close[stock] - w20_stock_close[stock].shift(1)) / w20_stock_close[stock].shift(1)
w20_stock_close
```
