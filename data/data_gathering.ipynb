```python
import yfinance as yf
```

```python
# List of WIG20 stock
wig20_stock = [
    'ALR.WA', 'ALE.WA', 'BDX.WA', 'CDR.WA', 'CPS.WA',
    'DNP.WA', 'JSW.WA', 'KGH.WA', 'KTY.WA', 'KRU.WA',
    'LPP.WA', 'MBK.WA', 'OPL.WA', 'PEO.WA', 'PCO.WA', 
    'PGE.WA', 'PKN.WA', 'PKO.WA', 'PZU.WA', 'SPL.WA'
]
```

```python
# Download data
w20_stock = yf.download(wig20_stock, period="max", interval="1d")

# Rename columns
w20_stock.columns = ['_'.join(col).strip() for col in w20_stock.columns.values]
w20_stock
```

```python
w20_stock.to_csv('./data_files/w20_stock.csv')
```
